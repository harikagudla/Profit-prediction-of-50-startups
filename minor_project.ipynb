```python
#converting categorical data to numerical data and then performing multilinear regression
```

```python
x=df.iloc[:,0:4]
```

```python
states=pd.get_dummies(x['State'])#converts categorical data to numerical data
```

```python
states
```

```python
x=x.drop('State',axis=1)#here drop deletes the column named State
```

```python
x=pd.concat([x,states],axis=1)#conat is used to join or add g and states
#now all the columns of dataframe has numerical data
```

```python
x
```

```python
y=df['Profit']
```

```python
import seaborn as sns
```

```python
#create a ml model to predict the profit using these all input parameters
```

```python
#sklearn takes only numeric data
#train_test_split: for any given number of values
#75% of values of x for x_train
#25% values of x for x_test
#in the same way 75% values of y for y_train
#25% values of y for y_test
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression #importing the algorithm and fitting(training the machine) the model
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)#predicting the output
y_pred
```

```python
y_test #actual data
```

```python
df1=pd.DataFrame({'Actual including categorical data':y_test,'Predicted including categorical data':y_pred})#creating a new dataframe consisting of actual and predicted data
df1
```

```python
sns.regplot(x="Actual including categorical data",y="Predicted including categorical data",data=df1)#REGRESSION PLOT FOR ACTUAL AND PREDICTED DATA(Data Visualization)
```

```python
df1.plot(figsize=(20,10))
plt.show()
```

```python
df1.plot(figsize=(15,10),kind="bar")
```

```python
from sklearn.metrics import r2_score #evaluation or finding the accuracy
r2_score(y_test,y_pred)
```
